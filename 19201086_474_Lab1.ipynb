## 5.28

```python
def wave_packet(a,b):
    x,t=symbols('x t')
    express=exp(-(x - 3 * t) ** 2)*sin(3 * pi * (x - t))
    ret=express.subs({x:a,t:b})

    return ret.evalf()



xlist = np.linspace(-4, 4, 1000)
ylist=[]
for i in xlist:
    ylist.append(wave_packet(i, 0))

plt.plot(xlist, ylist)
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.show()
```

### After switching to y=cos(18 * pi * x) and using sympy

```python
def func(x_val):
    x=Symbol('x')
    expr=cos(18 * pi * x)
    ret=expr.subs(x,x_val)
    return ret.evalf()

x = np.linspace(0,2,20)
X = np.linspace(0,2,1000)
y=[]
Y=[]
for i in x:
    y.append(func(i))
for i in X:
    Y.append(func(i))
plt.plot(x,y)
plt.plot(X,Y)
plt.title('Plotting f(x)=cos(18*pi*x)')
plt.legend(['20 points', '2000 points'])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

#### The blue curve is less accurate and the orange one is more accurate as it has more points (desmos plot example given above)

## 5.30

```python
def viscosity(t_val):
    t=Symbol('t')
    exprr= 2.414e-5 * 10 ** (247.8 / (t + 273.15 - 140.0))
    ret=exprr.subs(t,t_val)
    return ret.evalf()

T = np.linspace(0, 100, 1000)
visc=[]
for i in T:
    visc.append(viscosity(i))
    
plt.plot(T, visc)
plt.ylabel('viscosity (Pa s)')
plt.xlabel('temperature (C)')
plt.show()
```

## 5.31

```python
def c(lamb):
    l=Symbol('l')
    expp=sqrt(9.81 * l / (2 * pi) *(1 + 7.9e-2 * 4 * pi ** 2 / (1000 * 9.81 * l ** 2)) *tanh(2 * pi * 50.00 / l))
    ret=expp.subs(l,lamb)
    return ret.evalf()

l = np.linspace(0.001, 0.1, 1000)
c_arr1=[]
for i in l:
    c_arr1.append(c(i))
plt.plot(l, c_arr1)
plt.xlabel('wavelength (m)')
plt.ylabel('speed (m/s)')
plt.show()

l = np.linspace(1, 2000, 1000)
c_arr2=[]
for i in l:
    c_arr2.append(c(i))
plt.plot(l, c_arr2)
plt.xlabel('wavelength (m)')
plt.ylabel('speed (m/s)')
plt.show()
```
